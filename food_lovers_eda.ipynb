```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.width", 1000)
pd.set_option("display.max_colwidth", 200)

# Plot settings
plt.rcParams["figure.figsize"] = (10, 6)

# Read dataset
df = pd.read_csv("food_dely.csv")

# Basic validation
print("Dataset loaded successfully.\n")
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

print("\nFirst 5 rows:")
print(df.head())
```

```python
# Basic info
print("INFO:\n")
print(df.info())

# Describe numerical columns
print("\nNUMERICAL SUMMARY:\n")
print(df.describe())

# Describe categorical columns
print("\nCATEGORICAL SUMMARY:\n")
print(df.describe(include='object'))

# Unique values check for key columns
print("\nUNIQUE VALUES CHECK:\n")
cols_to_check = ['city', 'establishment', 'price_range', 'delivery', 'rating_text']

for col in cols_to_check:
    print(f"\nColumn: {col}")
    print(df[col].value_counts().head(10))
```

```python
# Missing values count
missing = df.isnull().sum()

# Missing percentage
missing_percent = (missing / len(df)) * 100

# Combine into one table
missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_percent
})

# Sort by highest missing
missing_df = missing_df.sort_values(by='Missing %', ascending=False)

print(missing_df)
```
